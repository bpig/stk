```python
o24 = o2.query("date >= '2024'").reset_index().copy()
import pandas_ta as ta
o24['e13'] = ta.ema(o24.close, 13)
o24['e21'] = ta.ema(o24.close, 21)
o24['e70'] = ta.ema(o24.close, 70)
o24['e84'] = ta.ema(o24.close, 84)

o24[::-1]
```

```python
o24['golden'] = ta.cross(o24.e21, o24.e70, above=True)
```

```python
o24.query('golden==1')
```

```python
o24['fast'] = o24[['e21', 'e13']].min(axis=1)
o24['slow'] = o24[['e70', 'e84']].max(axis=1)
o24['diff'] = o24['fast'] - o24['slow']
def show():
    return o24.loc[:, ['date', 'e13', 'e21', 'e70', 'e84', 'diff', 'smooth','cal']][::-1]
```

```python
o24['smooth'] = o24['diff'].rolling(window=2, min_periods=1).mean()
show()
```

```python
#o24.query('abs(smooth) <= 0.02')
o24.query('abs(cal) > .01')
```

```python
o24[['date','diff']].plot(x='date', y='diff')
```

```python
o24['cal'] = o24['diff'].diff(periods=1) / o24['diff']
ss = o24[['date','cal','diff']]
 
ax = ss.plot(x='date', y='cal')
ss.plot(x='date', y='diff',ax=ax, secondary_y=True)
```
